### __1b2. Investigate and impute null values - df_application__

```python
def find_nulls(df):
    """
    Convenience function to count and visualize null values in respective home credit file
    
    Parameters
    -----
    df: Dataframe of home credit file
    
    Returns
    -----
    Sorted dataframe of null features and their respective count and percentages (out of entire dataset)    
    """
    nulls = df.isnull().sum()
    nulls_percentage = (df.isnull().sum() / df.shape[0])
    df_nulls = pd.concat([nulls,np.round(nulls_percentage*100,decimals = 2)],axis = 1,keys = ['nulls','nulls_percentage'])
    
    print(f'Columns Missing Data: {df_nulls.query("nulls != 0").shape[0]}')
    return df_nulls.query("nulls != 0").sort_values(by = 'nulls_percentage',ascending = False).reset_index()
```

#### Analyze relationship between target and remaining null values in df_application

EDA on remaining null columns to assess whether to delete/impute

```python
# Convenience function to plot distributions of feature / target relationships

def plot_target_feature_relationship(df, feature):
    """
    Plots a specified a feature and its relation with target variable (whether applicant defaulted on loan)
    
    Parameters
    -----
    df: dataframe of home credit file
    feature: feature that is being analyzed
    
    Returns
    -----
    Plot of feature and target distribution
    """
    df = df.reset_index()
    plt.figure(figsize = (8, 4))
    plt.style.use('fivethirtyeight')
    try:
        # plot repaid loans
        sns.kdeplot(df.loc[df['TARGET'] == 0, feature], label = 'target == 0')
        # plot loans that were not repaid
        sns.kdeplot(df.loc[df['TARGET'] == 1, feature], label = 'target == 1')

        # Label the plots
        plt.title(f'Distribution of {feature} by Target Value')
        plt.xlabel('%s' % feature); plt.ylabel('Density');
        plt.show()
    except ValueError:
        df_feature = df.groupby(['TARGET',feature]).size().reset_index()
        plt.xticks(rotation = 70)
        sns.barplot(x= df_feature[feature],y=df_feature[0],hue=df_feature['TARGET'])
        plt.ylabel('Count');
        # Label the plots
        plt.title(f'Distribution of {feature} by Target Value')
```

```python
# Plot feature / target relationships for features with high null percentages in TRAIN dataset

null_features_application = find_nulls(df_application).query("nulls_percentage >= 10")['index'].tolist() #Find null features where percentage of total set >= 10%

for feature in null_features_application:
    plot_target_feature_relationship(df_application,feature)
```

##### There are visible relationships with EXT_SOURCE columns and occupation_type. Keep these features

```python
# Create separate dataframe with only columns that contain null values

df_previous_nulls = df_previous.loc[:,df_previous.isnull().any(axis = 0)]

# Append SK_ID_CURR to merge with main file df_application
df_previous_nulls['SK_ID_CURR'] = df_previous['SK_ID_CURR']
```

```python
# Bring in only rows with NaN values

null_mask = df_previous_nulls_merged['AMT_DOWN_PAYMENT'].isna()

df_previous_nulls_merged.loc[null_mask].head()
```

```python
# Groupby values for each occupation

df_occupation_nulls = pd.DataFrame((df_previous_nulls_merged.drop_duplicates(subset = 'SK_ID_CURR')
                                    .groupby('OCCUPATION_TYPE')
                                    .agg('AMT_DOWN_PAYMENT')
                                    .size()))
```

```python
# Introduce column which contains number of null values for each occupation

df_occupation_nulls['nulls'] = (df_previous_nulls_merged.drop_duplicates(subset = 'SK_ID_CURR')
                                .loc[df_previous_nulls_merged['AMT_DOWN_PAYMENT'].isna()]
                                .groupby('OCCUPATION_TYPE')
                                .agg('AMT_DOWN_PAYMENT')
                                .size())
```

#### Replace outliers with null values.

```python
def nullify_outliers(df, feature):
    """
    Converts all values deemed as an outlier to null value
    
    Parameters
    -----
    df: dataframe of home credit file
    feature: feature that is being analyzed and imputed for outliers
    
    Returns
    -----
    Dataframe will be updated with null values for outliers
    """
    anomfeature = str(feature) + '_ANOMALY'
    quantile = df[feature].quantile(q = .99)
    df[anomfeature] = df[feature] >= quantile
    df[feature] = df[feature].apply(lambda x: np.where(x >= quantile, np.nan,x))
```

```python
for feature in application_outliers:
    nullify_outliers(df_application,feature)
```

```python
# Check relationships after removing outliers

for feature in application_outliers:
    plot_target_feature_relationship(df_application,feature)
```
