# **Home Credit Default Risk**  
This project attempts to model and predict the default risk of consumer loan applicants by analyzing and extracting data from several sources including bureau data as well as previous home credit applications and installments. I divided the work into the following steps / sections:

## 1. Exploratory Data Analysis 
#### 1a. Baseline Models
\- Run logistic regression / random forest baseline models to gauge expectations and get an initial impression on main application file and the target variable  
#### 1b. Clean Data  
\- Investigate and impute null values as well as anomalies across main application file and supplemental files
#### 1c. Feature Exploration
\- Explore features by analyzing correlations and relationship with target variable (default / no default)  
## 2. Feature Engineering
#### 2a. Supplemental Data 
\- Engineer features by grouping and flattening supplemental datasets
#### 2b. Application Data
\- Merge supplemental data with main application dataframe. Create additional features through a combination of manual calculations with domain knowledge as well as polynomial features 
## 3. Gradient Boosting Classifier
#### 3a. Train LightGBM  
\- Train LightGBM model using data and previously engineered features  
#### ***3b. Model Results / Analysis***  
\- ***Compare models and assess feature importances***

```python
import pandas as pd
import numpy as np
import pickle

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

from sklearn.model_selection import train_test_split, KFold, StratifiedKFold
from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve, confusion_matrix
from lightgbm import LGBMClassifier

from rq_config import project_3_path

import os
import warnings
import time

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
# Establish model / data directories

model_version = '1'
lightgbm_dir = os.path.join(project_3_path,'models/',f'lightgbm{model_version}/')
baseline_dir = os.path.join(project_3_path,'models/','baseline/')
pkl_dir = os.path.join(project_3_path,'data/','pkl/')
```

<a id='3b1'></a>

### __3b1. Load Baseline and Gradient Boosting Models__

#### Load Results for Logistic Regression and Random Forest baseline models

```python
# Load precision / recall curves 
pr_curve_log = pickle.load(open(baseline_dir + 'pr_curve_log.pkl','rb'))
pr_curve_rfc = pickle.load(open(baseline_dir + 'pr_curve_rfc.pkl','rb'))
precision_log, recall_log, threshold_log = pr_curve_log[0], pr_curve_log[1], pr_curve_log[2]
precision_rfc, recall_rfc, threshold_rfc = pr_curve_rfc[0], pr_curve_rfc[1], pr_curve_rfc[2]

# Load ROC curves
roc_curve_log = pickle.load(open(baseline_dir + 'roc_curve_log.pkl','rb'))
roc_curve_rfc = pickle.load(open(baseline_dir + 'roc_curve_rfc.pkl','rb'))
fpr_log, tpr_log, thresholds_log = roc_curve_log[0], roc_curve_log[1], roc_curve_log[2]
fpr_rfc, tpr_rfc, thresholds_rfc = roc_curve_rfc[0], roc_curve_rfc[1], roc_curve_rfc[2]

# Load AUC scores
auc_log = pickle.load(open(baseline_dir + 'log_auc.pkl','rb'))
auc_rfc = pickle.load(open(baseline_dir + 'rfc_auc.pkl','rb'))
```

#### Load LightGBM Model and respective data

```python
# Load LightGBM Classifier

lgb_model = pickle.load(open(lightgbm_dir + 'model.pkl','rb'))
```

```python
# Load train/test data

data_train = pickle.load(open(lightgbm_dir + 'data_train.pkl','rb'))
data_test = pickle.load(open(lightgbm_dir + 'data_test.pkl','rb'))
```

```python
# Load feature importances dataframe from LightGBM model

feature_importance = pd.read_pickle(lightgbm_dir + 'feature_importance.pkl')
```

```python
# Extract features from dataset

features = data_train.iloc[:,2:].columns.tolist()
```

```python
# Extract predictions on test set

test_x, test_y = data_test[features], data_test['TARGET']
test_pred = lgb_model.predict_proba(test_x)[:,1]
```

```python
# Pull respective curves and metrics from LGB model

precision_lgb, recall_lgb, threshold_lgb = precision_recall_curve(test_y,test_pred)

fpr_lgb, tpr_lgb, thresholds_lgb = roc_curve(test_y,test_pred)

auc_lgb = roc_auc_score(test_y,test_pred)
```

<a id='3b2'></a>

### __3b2. Model Comparison - Recall Curves__

```python
# Plot recall scores across various thresholds.

fig,ax = plt.subplots(figsize = (12,6))

plt.plot(threshold_log,recall_log[1:],lw=2,label = 'Logistic Regression')
plt.plot(threshold_rfc,recall_rfc[1:],lw=2,label = 'Random Forest')
plt.plot(threshold_lgb,recall_lgb[1:],lw=2,label = f'LightGBM v{model_version}')
plt.legend()
plt.xlabel('Probability Threshold')
plt.ylim([-.05,1.05])
plt.title('Recall - Threshold');
```

<a id='3b3'></a>

### __3b3. Model Comparison - ROC Curves__ 

```python
plt.subplots(figsize = (12,10))
plt.plot(fpr_rfc, tpr_rfc,lw=2,label = f'ROC Random Forest (AUC: {auc_rfc:.2f})')
plt.plot(fpr_log, tpr_log,lw=2,label = f'ROC Logistic Regression (AUC: {auc_log:.2f})')
plt.plot(fpr_lgb, tpr_lgb,lw=2,label = f'ROC LightGBM v{model_version} (AUC: {auc_lgb:.2f})')
plt.plot([0,1],[0,1],c='violet',ls='--',label = f'Luck (AUC: 0.50)')
plt.legend()
plt.xlim([-0.05,1.05])
plt.ylim([-0.05,1.05])

plt.title('RFC Performance vs Log')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')

print(f'Baseline Logistic Regression AUC: {auc_log:.6%}')
print(f'Baseline Random Forest AUC: {auc_rfc:.6%}')
print(f'LightGBM v{model_version} AUC: {auc_lgb:.6%}');
```

#### Some interesting original features to visualize:

```python
def plot_feature_gain(features):
    """
    Convenience function to plot feature importance gain
    
    Parameters
    -----
    List of features to be plotted
    
    Returns
    -----
    Horizontal barplot with respective features and importance gain values
    """
    
    # Initialize list with most important feature as reference / anchor for other features to be visualized
    lst = ['EXT_SOURCE_2 EXT_SOURCE_3 DAYS_BIRTH']

    for feature in features:
        lst.append(feature)
        
    # Filter out feature importance df with features we're interested in visualizing
    feature_df = feature_importance[feature_importance['feature'].isin(lst)]
    feature_df = feature_df.sort_values(by = 'importance_gain',ascending = False)
        
    # Plot
    plt.style.use('ggplot')
    fig = plt.figure(figsize=(12, 8))
    gs = gridspec.GridSpec(1, 2)
    # Plot Split importances
    sns.barplot(x='importance_gain', y='feature', data=feature_df)
    plt.title('Feature Importance - Gain', fontweight='bold', fontsize=25)
    plt.tight_layout()
    plt.yticks(size = 20)
    fig.subplots_adjust(top=0.93)

```

```python
interesting_features = ['EXT_SOURCE_2 EXT_SOURCE_3 DAYS_BIRTH','CODE_GENDER_F','CODE_GENDER_M','DAYS_EMPLOYED','AMT_INCOME','OWN_CAR_AGE','DAYS_BIRTH']
```

```python
plot_feature_gain(interesting_features)
```
