### __2a1. Functions / aggregation recipes to groupby and calculate both numeric and categorical features__

```python
# Some null values in data set are custom-coded as 'XNA', 'XAP', or '365243.0' - replace them with actual null values

def replace_nulls(df):
    """
    Replaces null values custom-coded by Home Credit ('XNA','XAP',etc) with np.NaN
    
    Parameters
    -----
    df: Dataframe of home credit dataset
    
    Returns
    -----
    - Updated dataframe
    """

    df.replace(to_replace = {'XNA': np.nan, 'XAP': np.nan, 365243.0 : np.nan}, value = None, inplace = True)
    return df
```

```python
# Given the large set of features for each dataset, the functions below will assist in engineering features by automatically extracting and aggregating by selected features

def aggregate_numerics(df,df_name,group_var):
    """
    Groups together all numeric features by unique applicant SK_ID and calculates statistics.
    
    Parameters
    -----
    df: Dataframe from home credit dataset
    df_name: Name of dataframe which will be included in the newly created statistic features
    group_var: Unique identifier for group/file
    
    Returns
    -----
    - Updated dataframe with new features and all rows aggregated
    """
    for col in df:
        if col != group_var and 'SK_ID' in col:
            df = df.drop(columns = col)
            
    group_ids = df[group_var]
    numeric_df = df.select_dtypes('number')
    numeric_df[group_var] = group_ids

    # Group by the specified variable and calculate the statistics
    agg = numeric_df.groupby(group_var).agg(['count', 'mean', 'max', 'min', 'sum']).reset_index()

    # Need to create new column names
    columns = [group_var]

    # Iterate through the variables names
    for var in agg.columns.levels[0]:
        # Skip the grouping variable
        if var != group_var:
            # Iterate through the stat names
            for stat in agg.columns.levels[1][:-1]:
                # Make a new column name for the variable and stat
                columns.append('%s_%s_%s' % (df_name, var, stat))

    agg.columns = columns
    return agg
```

```python
def aggregate_categoricals(df, group_var, df_name):
    """
    Groups together all categorical features by unique applicant SK_ID and calculates statistics.
    
    Parameters
    -----
    df: Dataframe from home credit dataset
    df_name: Name of dataframe which will be included in the newly created statistic features
    group_var: Unique identifier for group/file
    
    Returns
    -----
    - Updated dataframe with new features and all rows aggregated
    """
    
    # Select the categorical columns
    categorical = pd.get_dummies(df.select_dtypes('object'))

    # Make sure to put the identifying id on the column
    categorical[group_var] = df[group_var]

    # Groupby the group var and calculate the sum and mean
    categorical = categorical.groupby(group_var).agg(['sum', 'mean'])
    
    column_names = []
    
    # Iterate through the columns in level 0
    for var in categorical.columns.levels[0]:
        # Iterate through the stats in level 1
        for stat in ['count', 'count_norm']:
            # Make a new column name
            column_names.append('%s_%s_%s' % (df_name, var, stat))
    
    categorical.columns = column_names
    
    return categorical
```

```python
# Create empty list - append all categorical features across all datasets for use in model

categorical_features = []
```

#### Calculate interest rate of each application where applicable:

```python
df_previous[['SK_ID_CURR','AMT_ANNUITY','AMT_CREDIT','CNT_PAYMENT']].head()
```

```python
# Impute interest rate from the above features using min max algorithm:

def interest_rate(pmt,pv,n):
    min_rate = 0
    max_rate = 20
    while min_rate < max_rate - .001:
        mid_rate = (min_rate + max_rate) / 2
        r = mid_rate / 100
        guessed_pmt = pv * (r/ ((1 - (1 + r) ** -(n+.000001) )))
        if guessed_pmt > pmt:
            max_rate = mid_rate
        else:
            min_rate = mid_rate
    return np.round(mid_rate,decimals = 2)
```

```python
# Create new interest rate column / feature

df_previous['INTEREST_RATE'] = df_previous[['AMT_ANNUITY','AMT_CREDIT','CNT_PAYMENT']].apply(lambda x: interest_rate(*x),axis = 1)
```

```python
# Replace outliers / null rates with np.NaN

df_previous['INTEREST_RATE'] = df_previous['INTEREST_RATE'].replace([20,0],np.NaN)
```

```python
df_previous_numeric_agg = aggregate_numerics(df_previous,'previous','SK_ID_CURR')
```

### __2a5. Groupby / flatten previous home credit application data - df_pos__

```python
# Load point of sale dataset

df_pos = pd.read_pickle(pkl_dir + 'POS_CASH_balance.pkl')
```

```python
df_pos.head()
```

```python
df_pos.shape
```

```python
# Calculate separate feature determining whether the contract was ever completed

df_pos['is_contract_status_completed'] = df_pos['NAME_CONTRACT_STATUS'] == 'Completed'
```

```python
# Create new dataframe for custom calculated features in df_pos

pos_features = pd.DataFrame({'SK_ID_CURR': df_pos['SK_ID_CURR'].unique()})
```

```python
# Groupby unique applicant ID then create new feature which is the most recent future installment count

group_object = df_pos.groupby('SK_ID_CURR')['CNT_INSTALMENT_FUTURE'].last().reset_index()
group_object.rename(index=str,
                    columns={'CNT_INSTALMENT_FUTURE': 'pos_cash_remaining_installments'},
                    inplace=True)

# Append to pos_features

pos_features = pos_features.merge(group_object, on=['SK_ID_CURR'], how='left')
```

```python
# Similar to above, groupby unique applicant ID then create new feature which is the most recent status of the contract

group_object = df_pos.groupby(['SK_ID_CURR'])['is_contract_status_completed'].sum().reset_index()
group_object.rename(index=str,
                    columns={'is_contract_status_completed': 'pos_cash_completed_contracts'},
                    inplace=True)

# Append to pos_features

pos_features = pos_features.merge(group_object, on=['SK_ID_CURR'], how='left')
```

```python
pos_features.head()
```

```python
# Aggregate all numerical features from df_pos

df_pos_numeric_agg = aggregate_numerics(df_pos,'pos','SK_ID_CURR')
```

```python
# Aggregate all categorical features from df_pos

df_pos_categorical_agg = aggregate_categoricals(df_pos,'SK_ID_CURR','pos')
```

```python
# Append categorical features

for feature in df_pos_categorical_agg.columns.tolist():
    categorical_features.append(feature)
```

```python
# Merge aggregated features

df_pos_agg = pd.merge(df_pos_numeric_agg,df_pos_categorical_agg,on = 'SK_ID_CURR')

del df_pos_numeric_agg, df_pos_categorical_agg
gc.collect()
```

```python
# Merge with grouped features that were generated above

df_pos_agg = pd.merge(df_pos_agg,pos_features,on = 'SK_ID_CURR')
```

```python
df_pos_agg.head()
```

```python
df_pos_agg.shape
```

```python
df_pos_agg.to_pickle(pkl_dir + 'df_pos_agg.pkl')
```

### __2a7. Groupby / flatten previous home credit application data - df_cc_balance__

```python
# Load credit card balances from previous home credit applicants

df_cc_balance = pd.read_pickle(pkl_dir + 'credit_card_balance.pkl')
```

```python
df_cc_balance.head()
```

```python
df_cc_balance.shape
```

```python
# Calculate the number of installments by grouping SK_ID_PREV and SK_ID_CURR and appending to original data

df_cc_balance['NUM_INSTALLMENTS'] = (df_cc_balance.groupby(['SK_ID_PREV','SK_ID_CURR'],as_index = False)
                                     ['CNT_INSTALMENT_MATURE_CUM'].max()
                                     ['CNT_INSTALMENT_MATURE_CUM'])
```

```python
# Aggregate numerical features

df_cc_balance_numeric_agg = aggregate_numerics(df_cc_balance,'cc_balance','SK_ID_CURR')
```

```python
# Aggregate categorical features

df_cc_balance_categorical_agg = aggregate_categoricals(df_cc_balance,'SK_ID_CURR','cc_balance')
```

```python
# Append categorical features

for feature in df_cc_balance_categorical_agg.columns.tolist():
    categorical_features.append(feature)
```

```python
# Merge

df_cc_balance_agg = pd.merge(df_cc_balance_numeric_agg,df_cc_balance_categorical_agg,on = 'SK_ID_CURR')

del df_cc_balance_numeric_agg, df_cc_balance_categorical_agg
gc.collect()
```

```python
# Weigh values by recency using months_balance feature

wm = lambda x: np.average(x, weights=-1/df_cc_balance.loc[x.index, 'MONTHS_BALANCE'])
```

```python
df_cc_wa = df_cc_balance.groupby('SK_ID_CURR')['AMT_BALANCE'].agg(wm).reset_index()
```

```python
df_cc_wa['AMT_DRAWINGS_ATM_CURRENT_WA'] = df_cc_balance.groupby('SK_ID_CURR')['AMT_DRAWINGS_ATM_CURRENT'].agg(wm)   
```

```python
df_cc_wa['AMT_DRAWINGS_CURRENT_WA'] = df_cc_balance.groupby('SK_ID_CURR')['AMT_DRAWINGS_CURRENT'].agg(wm)   
```

```python
df_cc_balance_agg = pd.merge(df_cc_balance_agg,df_cc_wa,on = 'SK_ID_CURR')
```
